```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
data = iris.data
target = iris.target
kmean = KMeans(n_clusters=3,n_jobs=4)
kmean.fit(data)
cents = kmean.cluster_centers_
labs = kmean.labels_
print(cents)
```

```python
fig,axes = plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(data[:,0],data[:,1],c=target,cmap='gist_rainbow',edgecolor='k',s=150)
axes[1].scatter(data[:,0],data[:,1],c=labs,cmap='gist_rainbow',edgecolor='k',s=150)
axes[0].set_xlabel('Sepal Length',fontsize=18)
axes[0].set_ylabel('Sepal Width',fontsize=18)
axes[1].set_xlabel('Sepal Length',fontsize=18)
axes[1].set_ylabel('Sepal Width',fontsize=18)
axes[0].tick_params(direction='in',length=10,width=5,colors='k',labelsize=20)
axes[1].tick_params(direction='in',length=10,width=5,colors='k',labelsize=20)
axes[0].set_title('Actual',fontsize=25)
axes[1].set_title('Predicted',fontsize=25)
axes[1].scatter(cents[:,0],cents[:,1], s=200, c='b', marker='s')
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(data[:,2],data[:,3],c=target,cmap='gist_rainbow',edgecolor='k',s=150)
axes[1].scatter(data[:,2],data[:,3],c=labs,cmap='gist_rainbow',edgecolor='k',s=150)
axes[0].set_xlabel('Petal Length',fontsize=18)
axes[0].set_ylabel('Petal Width',fontsize=18)
axes[1].set_xlabel('Petal Length',fontsize=18)
axes[1].set_ylabel('Petal Width',fontsize=18)
axes[0].tick_params(direction='in',length=10,width=5,colors='k',labelsize=20)
axes[1].tick_params(direction='in',length=10,width=5,colors='k',labelsize=20)
axes[0].set_title('Actual',fontsize=25)
axes[1].set_title('Predicted',fontsize=25)
axes[1].scatter(cents[:,2],cents[:,3], s=200, c='b', marker='s')
plt.show()
```

```python
c = 0
n = len(cents)
for i in range(n):
    if target[i] == labs[i]:
        c = c+1
acc = (c/n)*100
print("Accuracy =",acc,"\b%")
```
