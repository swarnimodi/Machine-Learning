```python
import pandas as pd, numpy as np
weights = np.zeros((13,1))
reader = pd.read_csv('housing.csv',header=None)
feature = reader.values.tolist()
features = []
for i in range(506):
    x = feature[i][0]
    y = x.split(' ')
    y[:] = (j for j in y if j!='')
    features.append(y)
sales = [float(row[13]) for row in features]
sales = np.array(sales).reshape(506,1)
for i in features:
    del i[13]
for i in range(506):
    for j in range(13):
        features[i][j] = float(features[i][j])
features = np.array(features).reshape(506,13)
features = normalize(features)
sales = normalize(sales)
for i in range(100):
    weights = update1(features,sales,weights,0.001)
    mse = costfunc(features,sales,weights)
    print(mse)
```

```python
def normalize(features):
    for f in features.T:
        fmean = np.mean(f)
        frange = np.amax(f) - np.amin(f)
        f -= fmean
        f /= frange
    return features
```

```python
def costfunc(features,sales,weights):
    predictions = np.zeros((506,1))
    predictions = np.dot(features,weights)
    sqerror = (sales-predictions)**2
    return sqerror.sum()/1012.0
```

```python
def update1(features,sales,weights,lr):
    predictions = np.zeros((506,1))
    predictions = np.dot(features,weights)
    for i in range(13):
        x = features[:,i]
        diff = -np.dot(x,(sales-predictions))
        weights[i][0] -= (lr*np.mean(diff))
    return weights
```

```python
def update2(features,sales,weights,lr):
    predictions = np.zeros((506,1))
    predictions = np.dot(features,weights)
    error = sales - predictions
    gradient = np.dot(-features.T,error)
    gradient *= (lr/506)
    weights -= gradient
    return weights
```
